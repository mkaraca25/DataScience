```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## Data OlusturalÄ±m

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

##Â Ilk Grafik

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

##Â Eksenler ve Lejant

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')

plt.legend(('sin', 'cos'))
```

##Â Subplotlar

```python
plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('x ekseni')
plt.plot(x, np.sin(x))
plt.ylabel('y ekseni')


plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
```

## TasarÄ±msal DeÄiÅiklikler

```python
fig2 = plt.figure(figsize=(10, 4))  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('x ekseni')
plt.plot(x, np.sin(x))
plt.ylabel('y ekseni')


# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color='red', linestyle='dashed', linewidth=4);
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
```

## Gorsellerin Kaydedilmesi

```python
fig2.savefig('ilk_gorsel.png')
```

# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

To show the line plots, letâs first import the famous iris data set.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_length'].plot()
```

```python
iris.columns
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index)
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index, hue='species')
```

## Scatter Plot

```python
fig, ax = plt.subplots()

ax.scatter(iris['sepal_length'], iris['sepal_width'])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
from numpy.random import randn
from matplotlib import pyplot

x = 20 * randn(1000) + 100
y = x + (10 * randn(1000) + 50)

pyplot.scatter(x, y)
pyplot.show()
```

## Barplot

```python
iris.groupby(by='species').mean()['petal_length']
```

```python
data = iris.groupby(by='species').mean()['petal_length']
ax = sns.barplot(x=data.index, y=data.values)


ax.set_title('Petal Length Means') 
ax.set_xlabel('Species') 
ax.set_ylabel('Mean')
```
