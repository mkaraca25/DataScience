```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1=np.linspace(0,10,20)
numpyDizisi2=numpyDizisi1**2
```

```python
(benimFigur,benimEksen)=plt.subplots()
```

```python
(benimFigur,benimEksen)=plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g",alpha=0.4)
benimEksen.plot(numpyDizisi2,numpyDizisi1,"b")
```

```python
(benimFigur,benimEksen)=plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi1+3,"g",alpha=0.4,linewidth=1.0)
benimEksen.plot(numpyDizisi1,numpyDizisi1+2,"b",linewidth=1.0)
benimEksen.plot(numpyDizisi1,numpyDizisi1+4,"#C96F23",linestyle="--")
benimEksen.plot(numpyDizisi1,numpyDizisi1+5,"#C96F23",linestyle=":")
benimEksen.plot(numpyDizisi1,numpyDizisi1+6,"#000000",linestyle=":",marker="o",markersize=5,markerfacecolor="red")
```

```python
plt.scatter(numpyDizisi1,numpyDizisi2)
```
