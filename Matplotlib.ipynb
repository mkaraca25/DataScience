```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
YasListesi=[10,20,30,30,30,40,50,60,70,75]
KiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi=np.array(YasListesi)
numpyKiloListesi=np.array(KiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
```

```python
plt.plot(numpyKiloListesi,numpyYasListesi,"b")
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa göre degisim grafigi")
```

```python
numpyDizisi1=np.linspace(0,10,10)
numpyDizisi1
numpyDizisi2=numpyDizisi1**3
numpyDizisi2
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.2,0.2,0.4,0.4])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
plt.xlabel("X ekseni veri ismi")
plt.ylabel("Y ekseni veri ismi")
plt.title("Grafik Basligi")
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni veri ismi")
eksen1.set_ylabel("Y ekseni veri ismi")
eksen1.set_title("Ana Grafik Basligi")
eksen2.plot(numpyDizisi2,numpyDizisi1,"r")
eksen2.set_xlabel("X ekseni veri ismi")
eksen2.set_ylabel("Y ekseni veri ismi")
eksen2.set_title("küçük Grafik Basligi")
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
type(benimEksenler)
for eksen in benimEksenler:
  eksen.plot(numpyDizisi1,numpyDizisi2,"g")
  eksen.set_xlabel("X Ekseni")
  eksen.set_xlabel("Y Ekseni")
```

```python
yeniFigur=plt.figure()#dpi=100
yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi**3")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpyDizisi**2")
yeniEksen.legend(loc=9)
```

```python
yeniFigur.savefig("BenimFigur.png",dpi=200)#save
```
