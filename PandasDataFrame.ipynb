**GroupBy**

```python
maasSozlugu=pd.DataFrame({"Departman":["Yazilim","Yazilim","Pazarlama","Pazarlama","Hukuk","Hukuk"],
                          "Calisan Ismi":["Mayk","Ahmet","Mehmet","Burak","Fatih","Ali"],
                          "Maas":[100,150,200,300,400,500]})
maasDtaFrame=pd.DataFrame(maasSozlugu)
maasDtaFrame
```

```python
grupObjesi=maasDtaFrame.groupby("Departman")
```

```python
grupObjesi.count()
```

```python
grupObjesi.mean()
```

```python
grupObjesi.describe()
```

**Concatenation**

```python
Sozluk1={"Isim":["Ahmet","Mehmet","Zeynep","Atil"],
         "Spor":["Kosu","Yuzme","Kosu","Basketbol"],
         "Kalori":[100,200,300,400]}
dataFrame1=pd.DataFrame(Sozluk1,index=[0,1,2,3])
dataFrame1
```

```python
Sozluk2={"Isim":["Osman","Levent","Atlas","Fatma"],
         "Spor":["Kosu","Yuzme","Kosu","Futbol"],
         "Kalori":[100,200,250,350]}
dataFrame2=pd.DataFrame(Sozluk2,index=[4,5,6,7])
dataFrame2
```

```python
Sozluk3={"Isim":["Ayse","Mahmut","Duygı","Nur"],
         "Spor":["Kosu","Yuzme","Badminton","Tenis"],
         "Kalori":[400,500,100,200]}
dataFrame3=pd.DataFrame(Sozluk3,index=[8,9,10,11])
dataFrame3
```

```python
BirlestirmeDataFrame=pd.concat([dataFrame1,dataFrame2,dataFrame3],axis=0)
BirlestirmeDataFrame
```

Merge

```python
Sozluk4={"Isim":["Ayse","Mahmut","Duygu","Nur"],
         "Spor":["Kosu","Yuzme","Badminton","Tenis"]}
dataFrame4=pd.DataFrame(Sozluk4)
dataFrame4
```

```python
Sozluk5={"Isim":["Ayse","Mahmut","Duygu","Nur"],
         "Kalori":[100,200,300,400]}
dataFrame5=pd.DataFrame(Sozluk5)
dataFrame5
```

```python
MergeDataFrame=pd.merge(dataFrame4,dataFrame5,on="Isim")
MergeDataFrame
```
